```python
#....... Simple example .......
#Machine Learning works like this:
# Input (Study Hours) → Model Learns → Output (Marks)

# Import library
from sklearn.linear_model import LinearRegression

# Input data (Study Hours)
X = [[1], [2], [3], [4], [5]]

# Output data (Marks)
Y = [50, 55, 65, 70, 80]

# Create model
model = LinearRegression()

# Train model
model.fit(X, Y)

# Give new input (6 hours study)
new_input = [[6]]

# Predict output
prediction = model.predict(new_input)

print("Predicted Marks:", prediction[0])
```
